```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from sklearn.cluster import KMeans
```

```python
img=cv2.imread('/content/sunset1.jpg')

img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
```

```python
X=img.reshape(-1,3)
print(X.shape)
k=4
model=KMeans(k)
model.fit(X)
```

```python
centroids=model.cluster_centers_
colors=np.array(centroids,dtype='uint8')
print(colors)

plt.imshow(colors)
plt.show()
```

```python
#For Color Patching(selecting a patch/part inside a cluster to see which is it's dominant colour)

i=1
for color in colors:
  plt.subplot(1,k,i)
  i+=1
  mat=np.zeros((10,10,3), dtype='uint8')  #initialising a patch
  mat[:,:,:]=color                        #filling the patch with the obtained colour
  plt.imshow(mat)
  plt.axis("off")

plt.show()
```
