**Data Preparation**

```python
#Dropping less impact features
df.drop(columns=['RowNumber','CustomerId','Surname'], inplace=True)
```

```python
df.head()
```

```python
df['Geography']=df['Geography'].astype('category').cat.codes
```

```python
df['Gender']=df['Gender'].astype('category').cat.codes
```

```python
df.tail()
```

```python
X=df.drop(columns=['Exited']).values
y=df['Exited'].values
```

```python
X.shape
```

```python
y.shape
```

**Model Training**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, classification_report, f1_score
```

```python
basic_model = XGBClassifier()
basic_model.fit(X_train, y_train)
```

**Model Evaluation**

```python
def model_evaluation(model):
  print('Training Accuracy:', model.score(X_train, y_train))
  print('Testing Accuracy:', model.score(X_test, y_test))
  y_pred = model.predict(X_test)
  print('F1 Score:', f1_score(y_test, y_pred))
  print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
```

```python
model_evaluation(basic_model)
```

**Hyperparameter Tuning**

```python
from sklearn.model_selection import GridSearchCV
```

```python
xgb_params = {
    'n_estimators': [100, 200],
    'max_depth':[4,5]
}
```

```python
grid=GridSearchCV(XGBClassifier(), param_grid = xgb_params, scoring= 'accuracy', cv=10)
grid.fit(X_train, y_train)
```

```python
print('Best Score', grid.best_score_)
print('Best Params', grid.best_params_)
```

```python
#Passing best parameters to the model will improve the efficiency
model1 = XGBClassifier(n_estimators= 100,max_depth=4)
model1.fit(X_train, y_train)
```

**Model Evaluation**

```python
model_evaluation(model1)
```
